```python
#list of highly correlated variables

x=x[['Persons- literate']]
y=x[x['Persons- literate'] > 0.7]
array = ["Persons","Males","Females","Number of households","Males- Literate","Females- Literate",\
         "Total Educated","Below Primary","Primary","Middle","Matric/Higher Secondary/Diploma",\
        "Graduate and Above","0 - 4 years","5 - 14 years","15 - 59 years","60 years and above (Incl. A.N.S.)",\
        "Total workers","Main workers","Non-workers","Religeon#1 Population","Imp Town#2 Population",\
        "Imp Town#3 Population"]


main_data =new_data [["Persons","Males","Females","Number of households","Males- Literate","Females- Literate",\
         "Total Educated","Below Primary","Primary","Middle","Matric/Higher Secondary/Diploma",\
        "Graduate and Above","0 - 4 years","5 - 14 years","15 - 59 years","60 years and above (Incl. A.N.S.)",\
        "Total workers","Main workers","Non-workers","Religeon#1 Population","Imp Town#2 Population",\
        "Imp Town#3 Population","Persons- literate"]].dropna()

#divide data into training dataset and testing dataset
train_df, test_df  = np.split(main_data, [int(len(main_data) * .80)])
    
# Output the two dataframes to separate files
train_df.to_csv('/home/rock19/Desktop/Jupyter_Home/train_census.csv')
test_df.to_csv('/home/rock19/Desktop/Jupyter_Home/test_census.csv')

main_data.info()
```

```python
#creating training and testing dataset using highly correlated variables

X_train = train_df[["Persons","Males","Females","Number of households","Males- Literate","Females- Literate",\
         "Total Educated","Below Primary","Primary","Middle","Matric/Higher Secondary/Diploma",\
        "Graduate and Above","0 - 4 years","5 - 14 years","15 - 59 years","60 years and above (Incl. A.N.S.)",\
        "Total workers","Main workers","Non-workers","Religeon#1 Population","Imp Town#2 Population",\
        "Imp Town#3 Population"]].astype('float').astype('int')

Y_train = train_df["Persons- literate"].astype('float').astype('int')

X_test  = test_df[["Persons","Males","Females","Number of households","Males- Literate","Females- Literate",\
         "Total Educated","Below Primary","Primary","Middle","Matric/Higher Secondary/Diploma",\
        "Graduate and Above","0 - 4 years","5 - 14 years","15 - 59 years","60 years and above (Incl. A.N.S.)",\
        "Total workers","Main workers","Non-workers","Religeon#1 Population","Imp Town#2 Population",\
        "Imp Town#3 Population"]].astype('float').astype('int')

Y_test = test_df["Persons- literate"].astype('float').astype('int')
```

```python
"""
Below are some machine learning algorithms, predicting no. of people literate for the test dataset 
along with their corresponding accuracy. """
```

```python
# K-neighbours
#choose k by bootstrap method (odd to avoid tied votes)

knn = KNeighborsClassifier(n_neighbors = 1)

knn.fit(X_train, Y_train)

Y_pred = knn.predict(X_test)

print('accuracy cofficient: ')
1-np.mean(np.abs((Y_test - Y_pred) / Y_test))
```

```python
#Random Forest

random_forest = RandomForestClassifier(n_estimators=10)

random_forest.fit(X_train, Y_train)

Y_pred = random_forest.predict(X_test)

random_forest.score(X_train,Y_train)

print('accuracy cofficient: ')
1-np.mean(np.abs((Y_test - Y_pred) / Y_test))
```

```python
# Gaussian Naive Bayes

gaussian = GaussianNB()

gaussian.fit(X_train, Y_train)

Y_pred = gaussian.predict(X_test)

print('accuracy cofficient: ')
1-np.mean(np.abs((Y_test - Y_pred) / Y_test))
```
